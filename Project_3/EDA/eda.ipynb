```python
#Label encoding the connection state columns
'''
Logic for doing this:
1. assigning values to each connection state
    conn_state_OTH -> 0 #other
    conn_state_REJ -> 1 #reject
    conn_state_RSTO -> 2 #reset originator
    conn_state_RSTOSO -> 3 #reset originator and responder
    conn_state_RSTR -> 4 #reset responder
    conn_state_RSTRH -> 5 #reset responder hard
    conn_state_SO -> 6 #syn only
    conn_state_S1 -> 7 #syn 1
    conn_state_S2 -> 8 #syn 2
    conn_state_S3 -> 9 #syn 3
    conn_state_SF -> 10 #syn fin
    conn_state_SH -> 11 #syn fin ack
    conn_state_SHR -> 12 #syn fin ack reset
2. For each row, check which column has value 1, and assign the corresponding value to a new column 'conn_state_encoded'
3. Drop the original connection state columns
df['conn_state_encoded'] = -1  # Initialize with a default value

# Using the linear search method to iterate through rows and then columns
for index, row in df.iterrows():
    for col in df.columns:
        if row[col] == 1:
            df.at[index, 'conn_state_encoded'] = conn_state_mapping[col]
            break

# Drop the original connection state columns
df = df.drop(columns=conn_state_columns)
'''
```

```python
# Fix the column names to match the actual DataFrame columns
connection_state_Colms = ["conn_state_OTH",
"conn_state_REJ",
"conn_state_RSTO",
"conn_state_RSTOS0",  # Fixed from RSTOSO to RSTOS0
"conn_state_RSTR",
"conn_state_RSTRH",
"conn_state_S0",      # Fixed from SO to S0
"conn_state_S1",
"conn_state_S2",
"conn_state_S3",
"conn_state_SF",
"conn_state_SH",
"conn_state_SHR"]

# Now select the columns
df_connection_col = df[connection_state_Colms]
```

```python
df_connection_col.head(50)
```

```python
#creating the mapping dictionary
conn_state_mapping = {
    "conn_state_OTH": 0,   # other
    "conn_state_REJ": 1,   # reject
    "conn_state_RSTO": 2,  # reset originator
    "conn_state_RSTOS0": 3,# reset originator and responder
    "conn_state_RSTR": 4,  # reset responder
    "conn_state_RSTRH": 5, # reset responder hard
    "conn_state_S0": 6,    # syn only
    "conn_state_S1": 7,    # syn 1
    "conn_state_S2": 8,    # syn 2
    "conn_state_S3": 9,    # syn 3
    "conn_state_SF": 10,   # syn fin
    "conn_state_SH": 11,   # syn fin ack
    "conn_state_SHR": 12   # syn fin ack reset
}

```

```python
#applying the mapping to create a new column
df_connection_col['conn_state_encoded'] = -1  # Initialize with a default value\

for index, row in df_connection_col.iterrows():
    for col in df_connection_col.columns:
        if row[col] == 1:
            df_connection_col.at[index, 'conn_state_encoded'] = conn_state_mapping[col]
            break
```

```python
df_connection_col.drop(columns=connection_state_Colms, inplace=True)
```

```python
df_og_wiothout_conn = df.drop(columns=connection_state_Colms)
```

```python
#concatenate the new column to the original dataframe
df_final = pd.concat([df_og_wiothout_conn, df_connection_col['conn_state_encoded']], axis=1)
```
